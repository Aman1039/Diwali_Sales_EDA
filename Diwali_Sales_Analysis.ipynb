# Exploratory Data Analysis

### Gender

```python
# plotting a bar chart for Gender and it's count
plt.figure(figsize=(10, 5))
ax = sns.countplot(x = 'Gender',data = df,palette ='colorblind')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# plotting a bar chart for gender vs total amount
plt.figure(figsize=(10, 5))
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, palette ='colorblind')
plt.show()
```

*Insight: From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', palette = "colorblind" ) # hue is used to further know about a specific attribut in count. Here we wanted to see by Gender.

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(7, 3))
sns.barplot(x='Age Group', y='Amount', hue='Age Group', data=sales_age, palette='colorblind', legend=False)
plt.show()
```

*Insight :From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ax1 = sns.barplot(data = sales_state, x = 'State',y= 'Orders',palette ='colorblind')
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount', palette ="colorblind")
plt.show()
```

*Insight: From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

## Spendings ##

***Lets find out where does Unmarried Females and Males spend their money.***

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for unmarried females and males
unmarried_females = df[(df['Gender'] == 'F') & (df['Marital_Status'] == 0)]
unmarried_males = df[(df['Gender'] == 'M') & (df['Marital_Status'] == 0)]

# Aggregate total orders by Product Category
product_category_sales_F = unmarried_females.groupby('Product_Category', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
product_category_sales_M = unmarried_males.groupby('Product_Category', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

# Set figure size
fig, axes = plt.subplots(1, 2, figsize=(20, 6))  # Create two subplots

# Plot for Unmarried Females
sns.barplot(data=product_category_sales_F, x='Product_Category', y='Orders', palette="coolwarm", ax=axes[0])
axes[0].set_title("Top Product Categories Purchased by Unmarried Females")
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)

# Plot for Unmarried Males
sns.barplot(data=product_category_sales_M, x='Product_Category', y='Orders', palette="viridis", ax=axes[1])
axes[1].set_title("Top Product Categories Purchased by Unmarried Males")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)

# Adjust layout for better readability
plt.tight_layout()

# Show plot
plt.show()
```

***Lets find out Where does Married Males and Females spend their money.***

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for unmarried females and males
unmarried_females = df[(df['Gender'] == 'F') & (df['Marital_Status'] == 1)]
unmarried_males = df[(df['Gender'] == 'M') & (df['Marital_Status'] == 1)]

# Aggregate total orders by Product Category
product_category_sales_F = unmarried_females.groupby('Product_Category', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
product_category_sales_M = unmarried_males.groupby('Product_Category', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

# Set figure size
fig, axes = plt.subplots(1, 2, figsize=(20, 6))  # Create two subplots

# Plot for Unmarried Females
sns.barplot(data=product_category_sales_F, x='Product_Category', y='Orders', palette="coolwarm", ax=axes[0])
axes[0].set_title("Top Product Categories Purchased by Married Females")
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)

# Plot for Unmarried Males
sns.barplot(data=product_category_sales_M, x='Product_Category', y='Orders', palette="viridis", ax=axes[1])
axes[1].set_title("Top Product Categories Purchased by Married Males")
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=90)

# Adjust layout for better readability
plt.tight_layout()

# Show plot
plt.show()
```

Insight: As we can see irrespective of their marital status people are spending more on Clothing,Food,Electronics and Footwears. After marraige the order numbers have decresed by 500 orders in both Genders.

### Conclusions ###

The analysis reveals that females dominate the buyer demographic, both in terms of numbers and purchasing power. The majority of buyers belong to the 26-35 age group, with married women showing the highest purchasing capacity. Geographically, Uttar Pradesh, Maharashtra, and Karnataka contribute the most to total sales and order volume.

Industry-wise, most buyers are employed in IT, Healthcare, and Aviation, indicating a strong purchasing trend among professionals. The most popular product categories include Food, Clothing, and Electronics, with Footwear also being a significant preference. Regardless of marital status, consumers prioritize spending on these categories. However, post-marriage, there is a noticeable decline of approximately 500 orders across both genders.

### Recommendations ###

To maximize sales, businesses should target female consumers, especially married women, with personalized discounts and loyalty programs, focusing on the 26-35 age group through social media and influencer marketing. Expanding in high-sales states like Uttar Pradesh, Maharashtra, and Karnataka with better logistics can boost revenue. Prioritizing Food, Clothing, Electronics, and Footwear while catering to working professionals in IT, Healthcare, and Aviation with convenience-driven products is crucial. To retain post-marriage customers, introducing family-centric deals and subscription plans can help.
